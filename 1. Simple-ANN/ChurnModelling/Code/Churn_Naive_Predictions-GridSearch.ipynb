```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('../Data/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:,13]
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
label_encode_country = LabelEncoder()
label_encode_Gender = LabelEncoder()
```

```python
X[:,1] = label_encode_country.fit_transform(X[:,1])
X[:,2] = label_encode_Gender.fit_transform(X[:,2])
```

```python
onehot = OneHotEncoder(categorical_features=[1])
```

```python
X = onehot.fit_transform(X).toarray()
```

```python
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
from keras.layers import Dropout
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    model = Sequential()
    model.add(Dense(units=6,kernel_initializer='uniform',activation='tanh',input_dim=11))
    model.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
    model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=build_classifier)
```

```python
parameters = {
    'batch_size':[25,32],
    'nb_epoch': [100,500],
    'optimizer' : ['adam','rmsprop'],
}
```

```python
grid_search = GridSearchCV(estimator=model,param_grid=parameters, scoring='accuracy',cv=10,n_jobs=-1)
```

```python
grid_search = grid_search.fit(x_train,y_train)
```

```python
best_parametres = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
